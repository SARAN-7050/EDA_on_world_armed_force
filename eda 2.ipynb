```python
import matplotlib.pyplot as plt
top_5 = df['Paramilitary'].sort_values(ascending=False).head(5)

countries = ['China', 'India', 'United States', 'North Korea', 'Russia']

plt.bar(countries, top_5)

plt.xlabel('Countries')
plt.ylabel('Paramilitary ')
plt.title('Top 5 Paramilitary Countries')
plt.grid(axis='y',alpha=0.3)

plt.show()
```

```python
import matplotlib.pyplot as plt
top_5 = df['Total'].sort_values(ascending=False).head(5)

countries = ['Gambia', 'europe', 'United States', 'North Korea', 'Russia']

plt.bar(countries, top_5)

plt.xlabel('Paramilitarry')
plt.ylabel('Total ')
plt.title('Top 5 Total Countries')
plt.grid(axis='x',alpha=0.2)

plt.show()
```

```python
import matplotlib.pyplot as plt
top_5 = df['Active military'].sort_values(ascending=False).head(5)

countries = ['China', 'India', 'United States', 'North Korea', 'Russia']

plt.bar(countries, top_5)

plt.xlabel('Countries')
plt.ylabel('Active military')
plt.title('Top 5 Active military Countries')
plt.grid(axis='y',alpha=0.3)

plt.show()
```
